```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df=pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['address','url','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
#How to droppng duplicates
df.shape
df.drop_duplicates(inplace=True)
df.shape
```

```python
#cleaning the each columns
```

```python
df['rate'].unique()
```

```python
## Remove stringg and other value from rate column
def handlerate(data):
    if data=='NEW' or data=='-':
        return np.nan
    else:
        data=str(data).split("/")
        data=data[0]
        return float(data)
```

```python
df['rate']=df['rate'].apply(handlerate)
df['rate'].unique()
```

```python
#how to handle null values
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2Plates','listed_in(type)':'type'},inplace=True)
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
df['Cost2Plates'].unique()
```

```python
def handlecomma(data):
    data=str(data)
    if ',' in data:
        return float(data.replace(",",""))
    else:
        return float(data)
    
```

```python
df['Cost2Plates']=df['Cost2Plates'].apply(handlecomma)
```

```python
df['Cost2Plates'].unique()
```

```python
df.head()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
```

```python
rest_types_less1000=rest_types[rest_types<1000]
```

```python
rest_types_less1000
```

```python
def handle_rest_type(data):
    if (data in rest_types_less1000):
        return 'others'
    else:
        return data
    

```

```python
df['rest_type']=df['rest_type'].apply(handle_rest_type)
```

```python
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_less100=cuisines[cuisines<100]
cuisines_less100
```

```python

def handle_cuisine(value):
    if(value in cuisines_less100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisine)
df['cuisines'].value_counts()
```
